## DOCTOR'S PRECRIPTION


````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
````

````python

prescriptions.columns = prescriptions.columns.str.strip().str.lower().str.replace(" ", "_")
````

````python
prescriptions['dosage'] = pd.to_numeric(prescriptions['dosage'], errors='coerce')
prescriptions['frequency'] = pd.to_numeric(prescriptions['frequency'], errors='coerce')
prescriptions['duration'] = pd.to_numeric(prescriptions['duration'], errors='coerce')
````

````python
prescriptions['disease'] = prescriptions['disease'].astype('category').cat.codes
prescriptions['drug'] = prescriptions['drug'].astype('category').cat.codes
````

````python

print(prescriptions.info())
````

````python
desc = prescriptions.describe().T[['mean', 'std', 'min', 'max']]
desc.columns = ['Mean', 'StdDev', 'Min', 'Max']
print(desc)
````

````python
plt.figure(figsize=(8, 6))
sns.heatmap(prescriptions.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
````

````python
patients.columns = patients.columns.str.strip().str.lower().str.replace(" ", "_")
````

````python
patients['gender'] = pd.to_numeric(patients['gender'].astype('category').cat.codes, errors='coerce')
patients['condition'] = pd.to_numeric(patients['condition'].astype('category').cat.codes, errors='coerce')
patients['procedure'] = pd.to_numeric(patients['procedure'].astype('category').cat.codes, errors='coerce')
patients['readmission'] = pd.to_numeric(patients['readmission'].astype('category').cat.codes, errors='coerce')
patients['outcome'] = pd.to_numeric(patients['outcome'].astype('category').cat.codes, errors='coerce')
````

````python
numeric_cols = patients.select_dtypes(include=['float64', 'int64']).columns
patients[numeric_cols] = patients[numeric_cols].astype(float)
````

````python
def short_describe(df):
    desc = df.describe().T[['mean', 'std', 'min', 'max']]
    desc.columns = ['Mean', 'StdDev', 'Min', 'Max']
    return desc
````

````python
print(patients.info())
print(short_describe(patients))
````

````python
plt.figure(figsize=(10, 6))
sns.heatmap(patients.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
````
